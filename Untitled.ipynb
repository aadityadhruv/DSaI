```python
#Required imports
import datetime as dt
import praw
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request
import os
import shutil
import cv2
import ProcessImages as pci
```

```python
cwd = os.getcwd()
datasets = "carts_data"
directory = os.path.join(cwd, datasets)
```

```python

for filename in os.listdir(directory):
    if filename.endswith(".png"): 
        print(filename)
        image = cv2.imread(datasets + "/" + filename)
        edges = pci.get_edges(image)
        x,y,w,h = pci.get_bounding_rect(edges, image)
        pci.draw_bounding_rect(x,y,w,h, datasets + "/" + filename)
        continue
    else:
        continue
```
